```python
retail_raw = pd.read_excel('online_retail.xlsx')
```

```python
retail_raw.head()
```

```python
retail = retail_raw.copy()
```

```python
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
```

```python
retail.isnull().sum()
```

```python
retail = retail[retail.Quantity >= 0]
```

```python
retail.info()
```

```python
retail.CustomerID = retail.CustomerID.astype('object')
retail.info()
```

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.head()
```

```python
# SELECT * FROM retail;
retail
```

```python
# SELECT Quantity, UnitPrice, CustomerID FROM retail;

retail[['Quantity', 'UnitPrice', 'CustomerID']]
```

```python
# SELECT Quantity, UnitProce, CustomerID FROM retail LIMIT 10;
retail[['Quantity', 'UnitPrice', 'CustomerID']].head(10)
```

```python
retail[['Quantity', 'UnitPrice', 'CustomerID']][:10]
```

```python
# SELECT * FROM retail WHERE CustomerID=17850;
retail[(retail.Quantity==17850)]
```

```python
retail.query('CustomerID == "17850"')
```

```python
# SELECT * FROM retail WHERE CustomerID=17850 and UnitPrice > 5;
retail[(retail.CustomerID == '17850') & (retail.UnitPrice > 5)]
```

```python
retail.query('CustomerID == "17850" & UnitPrice > 5')
```

```python
# SELECT * FROM retail WHERE CustomerID=17850 or Country > France;
retail[(retail.CustomerID == '17850') | (retail.Country == 'France')]
```

```python
retail.query('CustomerID == "17850" | Country == "France"')
```

```python
# SELECT * FROM retail WHERE InvoceNo in not null;
retail[retail.InvoiceNo.notnull()]
```

```python
# SELECT CustomerID, count(*) FROM reatil GROUP BY Customer ID;
retail.groupby(by='CustomerID').size()
```

```python
# SELECT CustomerID, avg(Revenue), count(*) FROM reatil GROUP BY Customer ID;
retail['Revenue'] = retail.Quantity * retail.UnitPrice
retail.groupby(by='CustomerID').agg({'Revenue': np.mean, 'CustomerID': np.size})
```

```python
retail.groupby(by='CustomerID').agg({'Revenue': np.mean, 'CustomerID': np.size}).rename(columns={'Revenue': 'RevenueAvg', "CustomerID": 'CustIDCount'})
```

```python
# SELECT InvoceDateDay sum(Revenue), count(*) FROM reatil GROUP BY InvoceDateDay;
retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day
result = retail.groupby(by='InvoiceDateDay').agg({'Revenue': np.sum})
```

```python
result.plot(kind='bar')
```
